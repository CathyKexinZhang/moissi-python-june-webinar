## Tabular Data
Processing tabular data using numpy

```python
import numpy
import os
```

```python
help(numpy.genfromtxt)
```

```python
distance_file = os.path.join('data', 'distance_data_headers.csv')
# delimiter: separate from coma
# dtype = 数据类型 keep all things as string/float
distances = numpy.genfromtxt(fname = distance_file, delimiter = ',', dtype = 'unicode')
print(distances)
```

```python
headers = distances[0]
print(headers)
```

```python
# Slice distances to get the numerical values (everything but the header)
data = distances[1:]
print(data)
```

```python
# scientific notation with float as data type
data = data.astype(numpy.float)
print(data)
```

```python
# To slice a 2D
# array_name[row_index, column_index]
print(data[0,1])
print(data[1,0])
```

```python
small_data = data[0:10, 0:3]
print(small_data)
```

```python
print(small_data[:, 1:])
```

```python
# data_average = numpy.mean(data_set)
thr4_atp = (data[:, 1])
avg_thr4_atp = numpy.mean(thr4_atp)
print(avg_thr4_atp)
```

```python
# Create a variable called num_columns which is the number of columns
num_columns = len(headers)
print(num_columns)
```

```python
# range(begin, end)
for i in range(1, num_columns):
    column = data[:, i]
    avg_col = numpy.mean(column)
    data_label = headers[i]
    print(F'{data_label}: {avg_col: .3f}')
```

## Tabular Data
Processing tabular data using numpy

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
plt.figure()
sample_name = headers[1]
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:,1],label = sample_name)
plt.legend()

plt.savefig(F'{sample_name}.png',  dpi = 300)
```

```python
plt.figure()
sample_name = headers[1]
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:, 1], label = sample_name, color = 'red')
plt.plot(data[:, 2], label = headers[2], color = 'blue')
plt.legend()

plt.savefig(F'{sample_name, headers [2]}',  dpi = 300)
```

```python
plt.figure()
for col in range(1, num_columns):
    plt.plot(data[: , col], label = headers[col])
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()
plt.savefig('all_samples.png')
```

```python
data_min = numpy.min(data[:, 1:])
data_max = numpy.max(data[:, 1:])

for col in range(1, num_columns):
    plt.figure()
    sample_name_each = headers[col]
    
    plt.plot(data[:, col], label = sample_name_each)
    plt.ylim(data_min, data_max)
    
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    plt.savefig(F'{sample_name_each}',  dpi = 300)
```
